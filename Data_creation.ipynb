```python
### Data 생성 

import pandas as pd
import os

# 1. 폴더 생성
os.makedirs("data", exist_ok=True)

# 2. 초기 데이터 (v1) 생성: iris 데이터 일부만 사용
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# 처음 100개 행만 저장 (v1 버전 시뮬레이션)
df.iloc[:100].to_csv("data/data.csv", index=False)
print("✅ v1 데이터 생성 완료: data/data.csv (100행)")
```
